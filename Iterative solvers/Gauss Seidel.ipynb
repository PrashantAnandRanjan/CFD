```python
import numpy as np

nx = 3
x = np.zeros(nx)
a = np.array([[3, 1, -2], [-1, 4, -3], [1, -1, 4]])
b = np.array([9, -8, 1])
e0 = .00009
e = np.zeros(nx)
xn = x.copy()
flag = 1
while flag <= 3:
    for i in range(nx):
        x[i] = (b[i] - np.sum(x[:] * a[i,:]) + xn[i] * a[i][i]) / a[i][i]
        print(x)
    e[:] = abs((x[:] - xn[:]) / x[:])
    print('\nError\t',e,'\n')
    xn = x.copy()
    for i in range(nx):
        if e[i] < e0:
            print('LLLLLL')
            flag+=1
```
